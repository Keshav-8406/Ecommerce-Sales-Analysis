This project analyzes sales data using Pandas to identify trends across regions, categories, and salespersons.

```python
# Dataset Cell

import pandas as pd
import random

random.seed(42)

regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)

print("First 10 Rows")
print(df.head(10))

print("\nDataset Shape:", df.shape)

print("\nDataset Info")
print(df.info())
```

## Task 1: Basic Grouping and Single Aggregations

In this task, the dataset is grouped to analyze total sales by region, number of transactions by category, and average sales by salesperson.

```python
# Total sales by region
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales = region_sales.sort_values(by='sales_amount', ascending=False)

# Transactions per category
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

# Average sales per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print("Total Sales by Region")
print(region_sales)

print("\n---------------------------------\n")

print("Transactions Per Category")
print(category_transactions)

print("\n---------------------------------\n")

print("Average Sales per Salesperson")
print(salesperson_avg)
```

## Task 2: Multi Column Grouping

This task performs advanced grouping using multiple columns and calculates multiple metrics to evaluate performance.

```python
# Sales by region and category
region_category_sales = df.groupby(['region','category'])['sales_amount'].sum().reset_index()

# Salesperson metrics
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum','mean','count']).reset_index()
salesperson_metrics = salesperson_metrics.sort_values(by='sum', ascending=False)

# Top category
category_sales = df.groupby('category')['sales_amount'].sum()
top_category = category_sales.idxmax()

# Top salesperson
top_salesperson = salesperson_metrics.iloc[0]['salesperson']

print("Sales by Region and Category")
print(region_category_sales)

print("\n---------------------------------\n")

print("Salesperson Performance")
print(salesperson_metrics)

print("\nTop Performing Salesperson:", top_salesperson)

print("\n---------------------------------\n")

print("Top Category:", top_category)
```

## Task 3: Custom Aggregation and Final Report

A custom function is used to calculate sales range. A final summary report is created to show total sales, average sales, and total transactions.

```python
def sales_range(x):
    return x.max() - x.min()

# Region analysis
region_analysis = df.groupby('region')['sales_amount'].agg(['sum','mean','min','max',sales_range]).reset_index()

# Final report
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

# Flatten columns
final_report.columns = ['region', 'sales_sum', 'sales_mean', 'customer_count']

print("Region Analysis")
print(region_analysis)

print("\n---------------------------------\n")

print("Final Report")
print(final_report)
```

## Explanation of Multi-Level Columns
**The final report creates multi-level columns because multiple aggregations are applied on the same column.
This results in hierarchical column names like sales_amount → sum and mean.
This structure provides more detailed insights compared to single aggregation.**
